```python
import time 
from bs4 import BeautifulSoup
import requests
from fake_useragent import UserAgent

# Background on user-agents

ua = UserAgent()

header = {'user-agent':ua.chrome}
url='https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=0&as-type=HISTORY&as-backfill=on'
# Getting the webpage, creating a Response object.
response = requests.get(url,headers=header)

# Extracting the source code of the page.
data = response.text

# Passing the source code to Beautiful Soup to create a BeautifulSoup object for it.
soup = BeautifulSoup(data, 'lxml')

response
```

```python
# Extracting all the <a> tags into a list.
mobile=[]
price=[]
i=1

while len(mobile) < 10574:
    attr={'class':'_3wU53n'}
    m = soup.find_all('div',attrs=attr)

    # Extracting URLs from the attribute href in the <a> tags.
    for tag in m:
        mobile.append(tag.string)

    att={'class':'_1vC4OE _2rQ-NK'}
    r = soup.find_all('div',attrs=att)
    for tag in r:
        price.append(tag.string)

    at={'class':'_3fVaIS'}
    ur = soup.find_all('a',attrs=at)
    if len(ur)>0:
        url='https://www.flipkart.com'+ur[0]['href']
    else:
        break
    time.sleep(3)
    response = requests.get(url,headers=header)
    data = response.text
    soup = BeautifulSoup(data, 'lxml')
    print(str(i)+" "+str(len(mobile)))
    i+=1
```
